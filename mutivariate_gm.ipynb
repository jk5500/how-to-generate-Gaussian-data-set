```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N=500
m1=0.4
m2=0.4
m3=0.2
mu1=[2,2]
mu2=[-2,2]
mu3=[5.5,2]
s1=[[1,0.1],[0.1,1]]
s2=[[1,-0.1],[-0.1,1]]
s3=[[1,0.2],[0.2,0.5]]
```

```python
X1=np.random.multivariate_normal(mu1,s1,[int(N*m1)])
X2=np.random.multivariate_normal(mu2,s2,[int(N*m2)])
X3=np.random.multivariate_normal(mu3,s3,[int(N*m3)])
```

```python
plt.plot(X1[:,0],X1[:,1],'rx',label='class 1')
plt.plot(X2[:,0],X2[:,1],'bx',label='class 2')
plt.plot(X3[:,0],X3[:,1],'gx',label='class 3')
plt.legend()
plt.grid()
plt.show()
```

```python
h=0.2
x_min, x_max = -6,10
y_min, y_max = -1,5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z=np.c_[xx.ravel(),yy.ravel()]
m=(xx.ravel()).shape
m=int(np.array(m))
```

```python
all_value1=[]
all_value2=[]
all_value3=[]
for i in range(m):
    value1=(np.matrix(Z[i]-mu1))*np.linalg.inv(s1)*(np.matrix(Z[i]-mu1)).T
    all_value1.append(value1)
    value2=(np.matrix(Z[i]-mu2))*np.linalg.inv(s2)*(np.matrix(Z[i]-mu2)).T
    all_value2.append(value2)
    value3=(np.matrix(Z[i]-mu3))*np.linalg.inv(s3)*(np.matrix(Z[i]-mu3)).T
    all_value3.append(value3)


```

```python
all_value1=np.squeeze(all_value1)
all_value2=np.squeeze(all_value2)
all_value3=np.squeeze(all_value3)
```

```python
Z1 = all_value1.reshape(xx.shape)
plt.contour(xx,yy, Z1,levels=[1,2,3,4])
Z2 = all_value2.reshape(xx.shape)
plt.contour(xx,yy, Z2,levels=[1,2,3,4])
Z3 = all_value3.reshape(xx.shape)
plt.contour(xx,yy, Z3,levels=[1,2,3,4])
plt.plot(X1[:,0],X1[:,1],'rx',label='class 1')
plt.plot(X2[:,0],X2[:,1],'bx',label='class 2')
plt.plot(X3[:,0],X3[:,1],'gx',label='class 3')
plt.legend()
plt.grid()
plt.show()
```
